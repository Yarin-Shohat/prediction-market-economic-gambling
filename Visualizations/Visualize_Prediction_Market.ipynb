```python
import plotly.express as px
import pandas as pd

# 1. Function to determine dominance
def get_dominance(row, gamb_col, econ_col):
    if row[gamb_col] == 1:
        return 'Gambling Dominant'
    elif row[econ_col] == 1:
        return 'Economic Dominant'
    else:
        return 'Equal/Neither'

# 2. Data preparation and "weight" calculation for each row (to get percentage of total data)
total_rows = len(data)
weight = (1 / total_rows) * 100

# Processing for Claude
df_claude = data.copy()
df_claude['Model'] = 'Claude'
df_claude['Dominance'] = df_claude.apply(lambda r: get_dominance(r, 'claude_combined_gambling_grater_than_economic', 'claude_combined_economic_grater_than_gambling'), axis=1)
df_claude['Percentage_Weight'] = weight

# Processing for GPT
df_gpt = data.copy()
df_gpt['Model'] = 'GPT'
df_gpt['Dominance'] = df_gpt.apply(lambda r: get_dominance(r, 'gpt_combined_gambling_grater_than_economic', 'gpt_combined_economic_grater_than_gambling'), axis=1)
df_gpt['Percentage_Weight'] = weight

# Concatenating into one final table
df_final = pd.concat([df_claude, df_gpt])
```

```python
import plotly.express as px
import pandas as pd

# 1. Calculate total percentages for each category (to add to the X-axis label)
category_totals = df_final.groupby('Classification')['Percentage_Weight'].sum() / 2 # Divide by 2 because we have the same rows twice (GPT and Claude)
category_labels = {cat: f"{cat}<br>({val:.1f}%)" for cat, val in category_totals.items()}

# 2. Update data with the new names
df_plot = df_final.groupby(['Model', 'Classification', 'Dominance'])['Percentage_Weight'].sum().reset_index()
df_plot['Classification_Label'] = df_plot['Classification'].map(category_labels)

# 3. Define color dictionary
colors = {
    'Gambling Dominant': '#ff7f0e', # Orange
    'Economic Dominant': '#a9a9a9', # Gray
    'Equal/Neither': '#1f77b4'      # Blue
}

# 4. Create the plot
fig = px.bar(
    df_plot,
    x="Classification_Label",
    y="Percentage_Weight",
    color="Dominance",
    facet_col="Model",
    title="Category Distribution & Dominance (Percentages in Labels)",
    # Order categories according to the new names
    category_orders={
        "Classification_Label": [category_labels['PRO'], category_labels['AGAINST'], category_labels['UNCLEAR']],
        "Dominance": ["Equal/Neither", "Gambling Dominant", "Economic Dominant"]
    },
    color_discrete_map=colors,
    labels={'Percentage_Weight': 'Percent of Total (%)', 'Classification_Label': ''},
    height=800
)

# 5. Clean design without text on the bars
fig.update_traces(
    marker_opacity=1,
    marker_line_width=0.5,
    marker_line_color='white'
)

# 6. Increase font sizes and style X-axis
fig.update_layout(
    barmode='stack',
    plot_bgcolor='white',
    font=dict(size=20, color="black"),
    title_font=dict(size=26),
    legend_title_text='Dominance Type',
    legend=dict(font=dict(size=18), orientation="h", y=1.1),
    margin=dict(t=150, b=100)
)

# 7. Dense grid lines
fig.update_yaxes(
    showgrid=True,
    gridcolor='LightGray',
    dtick=5,
    zeroline=True,
    zerolinecolor='black',
    tickfont=dict(size=18)
)

# 8. Bold model names at the bottom
fig.update_xaxes(row=1, col=1, title_text="<b>GPT Analysis</b>", title_font=dict(size=24, color="black"))
fig.update_xaxes(row=1, col=2, title_text="<b>Claude Analysis</b>", title_font=dict(size=24, color="black"))

# Remove Facet titles from the top
fig.for_each_annotation(lambda a: a.update(text=""))

fig.show()
```

```python
import plotly.express as px
import pandas as pd

# 1. Calculate total percentages for each category and each dominance within it
# Calculate average sums between both models for the X-axis label
summary_df = df_final.groupby(['Classification', 'Dominance'])['Percentage_Weight'].sum().reset_index()
summary_df['Percentage_Weight'] /= 2  # Average between GPT and Claude

# Create detailed labels dictionary
category_labels = {}
for cat in df_final['Classification'].unique():
    # Overall percentage of the category (PRO/AGAINST/UNCLEAR)
    total_pct = summary_df[summary_df['Classification'] == cat]['Percentage_Weight'].sum()

    # Extract percentages for each dominance
    g_pct = summary_df[(summary_df['Classification'] == cat) & (summary_df['Dominance'] == 'Gambling Dominant')]['Percentage_Weight'].sum()
    e_pct = summary_df[(summary_df['Classification'] == cat) & (summary_df['Dominance'] == 'Economic Dominant')]['Percentage_Weight'].sum()
    eq_pct = summary_df[(summary_df['Classification'] == cat) & (summary_df['Dominance'] == 'Equal/Neither')]['Percentage_Weight'].sum()

    # Build text string with line breaks (HTML <br>)
    label = (f"<b>{cat} ({total_pct:.1f}%)</b><br>"
             f"<span style='color:#ff7f0e'>G: {g_pct:.1f}%</span> | "
             f"<span style='color:#a9a9a9'>E: {e_pct:.1f}%</span> | "
             f"<span style='color:#1f77b4'>Eq: {eq_pct:.1f}%</span>")
    category_labels[cat] = label

# 2. Update data for plotting
df_plot = df_final.groupby(['Model', 'Classification', 'Dominance'])['Percentage_Weight'].sum().reset_index()
df_plot['Classification_Label'] = df_plot['Classification'].map(category_labels)

# 3. Define color dictionary
colors = {
    'Gambling Dominant': '#ff7f0e', # Orange
    'Economic Dominant': '#a9a9a9', # Gray
    'Equal/Neither': '#1f77b4'      # Blue
}

# 4. Create the plot
fig = px.bar(
    df_plot,
    x="Classification_Label",
    y="Percentage_Weight",
    color="Dominance",
    facet_col="Model",
    title="Category distribution and dominance (percentage of total responses)",
    category_orders={
        "Classification_Label": [category_labels['PRO'], category_labels['AGAINST'], category_labels['UNCLEAR']],
        "Dominance": ["Equal/Neither", "Gambling Dominant", "Economic Dominant"]
    },
    color_discrete_map=colors,
    labels={'Percentage_Weight': 'Percent of Total (%)', 'Classification_Label': ''},
    height=900
)

# 5. Style the appearance of the bars
fig.update_traces(
    marker_opacity=1,
    marker_line_width=0.5,
    marker_line_color='white'
)

# 6. Increase font sizes and style the Layout
fig.update_layout(
    barmode='stack',
    plot_bgcolor='white',
    font=dict(size=16, color="black"), # Slightly reduced to 16 to fit the complex details
    title_font=dict(size=24),
    legend_title_text='Dominance Type',
    legend=dict(orientation="h", yanchor="bottom", y=1.05, xanchor="right", x=1),
    margin=dict(t=180, b=150) # Increased bottom margin for detailed text
)

# 7. Grid lines
fig.update_yaxes(showgrid=True, gridcolor='LightGray', dtick=5)

# 8. Bold model names at the bottom
fig.update_xaxes(row=1, col=1, title_text="<b>GPT Analysis</b>", title_font=dict(size=22))
fig.update_xaxes(row=1, col=2, title_text="<b>Claude Analysis</b>", title_font_size=22)

# Remove Facet titles
fig.for_each_annotation(lambda a: a.update(text=""))

fig.show()
```

```python
import pandas as pd

# 1. Defining the mapping of column names for a clean table display
mapping = {
    'claude_combined_economic_ratio': 'Economic Theory Language (%)',
    'claude_combined_gambling_ratio': 'Gambling Language (%)',
    'claude_combined_Neutral_ratio': 'Ambiguous Language (%)',
    'claude_combined_economic_grater_than_gambling': 'Comments where Economic > Gambling',
    'claude_combined_gambling_grater_than_economic': 'Comments where Gambling > Economic'
}

def generate_classification_stats(df, classification_label, display_name):
    # Filtering the data by the Label in the Classification column
    subset = df[df['Classification'] == classification_label].copy()

    if subset.empty:
        print(f"No data found for classification: {classification_label}")
        return None

    # Ensure the "Greater Than" columns are in percentages (0 or 100) for the table average
    for col in mapping.keys():
        if col in subset.columns:
            # If the column is boolean or 0/1, convert it to 0/100
            if subset[col].max() <= 1:
                subset[col] = subset[col].astype(float) * 100

    # Selecting the relevant columns that exist in the DataFrame
    cols_to_stats = [col for col in mapping.keys() if col in subset.columns]

    # Calculating statistics
    stats_table = subset[cols_to_stats].agg(['count', 'mean', 'std', 'median', 'min', 'max']).T
    stats_table.columns = ['N', 'Mean', 'SD', 'Median', 'Min', 'Max']

    # Renaming the rows to the requested names
    stats_table = stats_table.rename(index=mapping)

    # Printing header and styling the table
    print(f"\n\033[1mOnly Comments {display_name} Proposed Rule\033[0m")
    return stats_table.round(2)

# 2. Running the analysis for each Label in the Classification column
labels_to_analyze = {
    'AGAINST': 'Against',
    'PRO': 'Pro',
    'UNCLEAR': 'Neutral to the'
}

for label, title_part in labels_to_analyze.items():
    stats_result = generate_classification_stats(data, label, title_part)
    if stats_result is not None:
        display(stats_result)
```
